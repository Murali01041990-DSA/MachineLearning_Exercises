```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return(lw,uw)
```

```python
wisker(df['GDP'])
```

```python
def z_score(col):
    zscore_col=(df[col]-df[col].mean())/df[col].std()
    return zscore_col
```

```python
z_score('GDP')
```

```python
for i in ['GDP','Total_expenditure','thinness__1-19_years','thinness_5-9_years']:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```

```python
for i in ['GDP','Total_expenditure','thinness__1-19_years','thinness_5-9_years']:
    sns.boxplot(data=df,x=i)
    plt.show()
```
