```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df_new=df
```

```python
df_new['Sex']=le.fit_transform(df_new[['Sex']])
df_new['ChestPainType']=le.fit_transform(df_new[['ChestPainType']])
df_new['RestingECG']=le.fit_transform(df_new[['RestingECG']])
df_new['ExerciseAngina']=le.fit_transform(df_new[['ExerciseAngina']])
df_new['ST_Slope']=le.fit_transform(df_new[['ST_Slope']])
```

```python
df_new.head(2)
```

```python
df_new.ExerciseAngina.unique()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df_new['Age']=scaler.fit_transform(df_new[['Age']])
df_new['ChestPainType']=scaler.fit_transform(df_new[['ChestPainType']])
df_new['RestingBP']=scaler.fit_transform(df_new[['RestingBP']])
df_new['Cholesterol']=scaler.fit_transform(df_new[['Cholesterol']])
df_new['ST_Slope']=scaler.fit_transform(df_new[['ST_Slope']])
df_new['MaxHR']=scaler.fit_transform(df_new[['MaxHR']])
```

```python
df_new.head(4)
```

```python
x=df_new.drop('HeartDisease',axis='columns')
y=df_new['HeartDisease']
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.svm import SVC
svm=SVC()
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10)
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2)
from sklearn.model_selection import cross_val_score
cvlr=cross_val_score(lr,xtrain,ytrain)
cvsvm=cross_val_score(svm,xtrain,ytrain)
cvrf=cross_val_score(rf,xtrain,ytrain)
```
